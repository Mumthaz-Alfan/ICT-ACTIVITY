```python
# import pandas ,numpy Task 1
import pandas as pd
import numpy as np

# load dataset using pd.read_csv()
df=pd.read_csv('/content/IMDB-Movie-Data.csv')
df.head()


```

```python


# 1. Display Shape (returns a tuple representing rows and columns)
print("--- 1. Dataset Shape ---")
print(df.shape)
print("\n" + "="*40 + "\n")

# 2. Display Head (shows the first 5 rows of data)
print("--- 2. First 5 Rows (Head) ---")

display(df.head())
print("\n" + "="*40 + "\n")

# 3. Display Column Names
print("--- 3. Column Names ---")
print(df.columns.tolist())
print("\n" + "="*40 + "\n")

# 4. Display Data Types
print("--- 4. Column Data Types ---")
print(df.dtypes)
print("\n" + "="*40 + "\n")

# 5. Display Missing Values
print("--- 5. Missing Values Count ---")
print(df.isnull().sum())
```

```python


# 1. Check for duplicates (returns the total count of duplicate rows)
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows before removal: {duplicate_count}")

# 2. Remove duplicates

df = df.drop_duplicates()


print(f"Number of duplicate rows after removal: {df.duplicated().sum()}")
```

```python


# Rename the column 'Title' to 'Movie_Title'
df = df.rename(columns={"Title": "Movie_Title"})

# Verify the change by listing the columns
print("Updated Column Names:")
print(df.columns.tolist())

# Show the head of the DataFrame to see the updated table header
df.head()
```

```python
#  Get unique genres


unique_genres = df['Genre'].str.split(',').explode().str.strip().unique()
print("Unique Genres in the Dataset:")
print(unique_genres)
print(f"\nTotal number of unique genres: {len(unique_genres)}")
```

```python
#  Extract number of genres per movie


df['Genre_Count'] = df['Genre'].apply(lambda x: len(x.split(',')))

df[['Movie_Title', 'Genre', 'Genre_Count']].head()
```

```python


# 1. Check missing values before handling
print("Missing values before:")
print(df[['Revenue (Millions)', 'Metascore']].isnull().sum())
print("\n" + "-"*40 + "\n")

# 2. Fill missing values with the mean of each column
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].mean())

# 3. Verify that there are no more missing values
print("Missing values after filling with mean:")
print(df[['Revenue (Millions)', 'Metascore']].isnull().sum())
```

```python
#  Top 5 movies by Rating
top_5_movies = df[['Movie_Title', 'Rating']].sort_values(by='Rating', ascending=False).head(5)

# Display the result
print("Top 5 Movies by Rating:")
display(top_5_movies)
```

```python
#  Movies with highest/lowest revenue

#  Movie with the highest revenue
highest_rev_idx = df['Revenue (Millions)'].idxmax()
highest_revenue_movie = df.loc[highest_rev_idx, ['Movie_Title', 'Revenue (Millions)']]

# Movie with the lowest revenue
lowest_rev_idx = df['Revenue (Millions)'].idxmin()
lowest_revenue_movie = df.loc[lowest_rev_idx, ['Movie_Title', 'Revenue (Millions)']]

# Display the results
print("--- Movie with the Highest Revenue ---")
print(f"Title: {highest_revenue_movie['Movie_Title']}")
print(f"Revenue: ${highest_revenue_movie['Revenue (Millions)']}M")
print("\n" + "-"*40 + "\n")
print("--- Movie with the Lowest Revenue ---")
print(f"Title: {lowest_revenue_movie['Movie_Title']}")
print(f"Revenue: ${lowest_revenue_movie['Revenue (Millions)']}M")
```

```python
#  Avg. rating by Director

# Group by Director and find the mean rating
avg_rating_by_director = df.groupby('Director')['Rating'].mean()

#  Display the first 10 directors in alphabetical order
print("--- Average Rating by Director (Alphabetical) ---")
print(avg_rating_by_director.head(10))

print("\n" + "="*40 + "\n")

#  Display the top 10 directors with the highest average rating
print("--- Top 10 Directors by Highest Avg Rating ---")
print(avg_rating_by_director.sort_values(ascending=False).head(10))
```

```python
#  Movies per year

# Count the number of movies released each year
movies_per_year = df['Year'].value_counts()

print("--- Number of Movies Per Year (Sorted by Highest Count) ---")
print(movies_per_year)

print("\n" + "="*40 + "\n")

#  Chronological order (Sorted by Year)
print("--- Number of Movies Per Year (Chronological Order) ---")
print(movies_per_year.sort_index())
```

```python
#  Movies with rating > 8 & revenue > 100M

# conditional filtering (both conditions must be true)
high_rated_high_revenue = df[(df['Rating'] > 8) & (df['Revenue (Millions)'] > 100)]

# Display specific columns for a cleaner view
result = high_rated_high_revenue[['Movie_Title', 'Rating', 'Revenue (Millions)']]

print(f"Total movies found matching criteria: {len(result)}")
display(result)
```

```python
#  Group by Director: avg Rating, Revenue


director_stats = df.groupby('Director').agg({
    'Rating': 'mean',
    'Revenue (Millions)': 'mean'
})

# Rename the columns slightly to make it clear they represent averages
director_stats = director_stats.rename(columns={
    'Rating': 'Avg_Rating',
    'Revenue (Millions)': 'Avg_Revenue'
})

# Display the top 10 rows sorted by highest average revenue
print("--- Director Stats Sorted by Highest Avg Revenue ---")
display(director_stats.sort_values(by='Avg_Revenue', ascending=False).head(10))
```

```python
#  Group by Year: count, avg Rating


year_stats = df.groupby('Year').agg({
    'Movie_Title': 'count',
    'Rating': 'mean'
})


year_stats = year_stats.rename(columns={
    'Movie_Title': 'Movie_Count',
    'Rating': 'Avg_Rating'
})

# Display the final summary sorted chronologically by year
print("--- Movie Count and Average Rating by Year ---")
display(year_stats.sort_index())
```

```python
#  Create a new column Profitability

# a) Calculate Profitability
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']

# b) Top 10 most profitable movies
top_10_profitable = df[['Movie_Title', 'Revenue (Millions)', 'Runtime (Minutes)', 'Profitability']].sort_values(by='Profitability', ascending=False).head(10)

print("--- Top 10 Most Profitable Movies ---")
display(top_10_profitable)
```

```python


def categorize_rating(rating):
    if rating < 6:
        return 'Poor'
    elif rating < 7.5:
        return 'Average'
    else:
        return 'Excellent'


df['Rating_Category'] = df['Rating'].apply(categorize_rating)


print("--- Sample Rows with the New Rating Categories ---")
display(df[['Movie_Title', 'Rating', 'Rating_Category']].head(10))

print("\n--- Breakdown of Rating Categories across all movies ---")
print(df['Rating_Category'].value_counts())
```

```python


# 1.  custom scoring function
def calculate_high_score(row):
    return (row['Rating'] * 10) + row['Revenue (Millions)']

# 2.  function to the dataset row-by-row (axis=1) to assign the score
df['High_Score'] = df.apply(calculate_high_score, axis=1)


print("--- Movies with Custom High Scores ---")
display(df[['Movie_Title', 'Rating', 'Revenue (Millions)', 'High_Score']].head())
```
