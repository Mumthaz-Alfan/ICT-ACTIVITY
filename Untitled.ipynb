```python


account_info = {
    "holder_name": "Mumtaz Noushad",
    "account_number": 16007658,
    "balance": 1000.00
}

def check_balance():

    print(f"\n--- Balance Inquiry ---")
    print(f"Account Holder: {account_info['holder_name']}")
    print(f"Current Balance: Rs{account_info['balance']:.2f}")

def deposit(amount):

    print(f"\n--- Deposit Operation ---")
    if amount > 0:
        account_info["balance"] += amount
        print(f"Successfully deposited: Rs{amount:.2f}")
        print(f"New Balance: Rs{account_info['balance']:.2f}")
    else:
        print("Error: Deposit amount must be greater than zero.")

def withdraw(amount):

    print(f"\n--- Withdrawal Operation ---")
    if amount <= 0:
        print("Error: Withdrawal amount must be greater than zero.")
        return


    if amount <= account_info["balance"]:
        account_info["balance"] -= amount
        print(f"Successfully withdrew: Rs{amount:.2f}")
        print(f"Remaining Balance: Rs{account_info['balance']:.2f}")
    else:
        print("Transaction Declined: Insufficient funds.")
        print(f"Current Available Balance: Rs{account_info['balance']:.2f}")


if __name__ == "__main__":

    check_balance()
    deposit(500.50)
    withdraw(300.00)
    withdraw(2000.00)
    check_balance()
```

```python


def determine_weather_condition(temperature, humidity):


    if humidity > 80:
        return "Rainy / Stormy"
    elif temperature >= 28 and humidity < 50:
        return "Sunny / Hot"
    elif 15 <= temperature < 28 and 50 <= humidity <= 80:
        return "Pleasant / Cloudy"
    elif temperature < 15:
        return "Cold / Chilly"
    else:
        return "Moderate / Variable"


def analyze_weather_data(region_name, temperature_readings, humidity_readings):



    print(f"=== Weather Analysis Report for Region: {region_name} ===")


    if not temperature_readings or not humidity_readings:
        print("Error: Missing weather reading metrics for this analysis.\n")
        return


    avg_temp = sum(temperature_readings) / len(temperature_readings)
    avg_humidity = sum(humidity_readings) / len(humidity_readings)


    overall_condition = determine_weather_condition(avg_temp, avg_humidity)


    print(f"Total Recorded Days  : {len(temperature_readings)}")
    print(f"Average Temperature  : {avg_temp:.2f}°C")
    print(f"Average Humidity     : {avg_humidity:.2f}%")
    print(f"Determined Condition : {overall_condition}")
    print("-" * 50 + "\n")




region_1 = "TRIVANDRUM"
temp_data_1 = [32.5, 34.0, 31.8, 35.2, 33.0]
humidity_data_1 = [25, 22, 28, 20, 24]


analyze_weather_data(region_1, temp_data_1, humidity_data_1)



region_2 = "THRISSUR"
temp_data_2 = [26.5, 25.0, 24.8, 26.0, 25.5]
humidity_data_2 = [85, 92, 89, 95, 90]


analyze_weather_data(region_2, temp_data_2, humidity_data_2)



region_3 = "KOLLAM"
temp_data_3 = [12.0, 14.5, 11.0, 9.5, 13.0]
humidity_data_3 =[34,77,87,65,69]
```
