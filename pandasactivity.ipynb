```python
import pandas as pd
import numpy as np


df = pd.read_csv("/titanic_dataset.csv")
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.groupby('Sex')['Survived'].mean() * 100
```

```python
df['Age'].agg(['mean', 'median', 'std'])
```

```python
df['Pclass'].value_counts()
```

```python
df[(df['Sex'] == 'female') & (df['Age'] < 18) & (df['Survived'] == 1)]
```

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
df['Age'].isnull().sum()  # Verifying 0 missing values remain
```

```python
conditions = [
    df['Age'] < 18,
    (df['Age'] >= 18) & (df['Age'] <= 60),
    df['Age'] > 60
]
choices = ['Child', 'Adult', 'Senior']

df['age_group'] = np.select(conditions, choices, default='Unknown')
df['age_group'].value_counts()
```

```python
df.nlargest(5, 'Fare')
```

```python
df.groupby(['Pclass', 'Sex'])['Fare'].mean()
```

```python
df = df.rename(columns={'Sex': 'gender', 'Fare': 'ticket_fare'})
df.columns.tolist()
```
